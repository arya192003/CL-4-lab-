# DATA EXTRACTION

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

print("# DATA EXTRACTION")
# Extract data from CSV
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)
print("Data loaded successfully")

```

```python
print(df.head())
```

# DATA LOADING

```python

print("\n# DATA LOADING")
# Load into cleaned DataFrame
cleaned_df = df.copy()
print("Data loaded into cleaned dataframe")
print(f"Shape of cleaned dataframe: {cleaned_df.shape}")
```

# DATA TRANSFORMATION

```python
# DATA TRANSFORMATION
print("\n# DATA TRANSFORMATION")

# 1. Handle missing values (no inplace, no warning)
cleaned_df['Age'] = cleaned_df['Age'].fillna(cleaned_df['Age'].median())
cleaned_df['Embarked'] = cleaned_df['Embarked'].fillna(cleaned_df['Embarked'].mode()[0])

# 2. Drop irrelevant or high-missing columns (ignore error if column is missing)
cleaned_df = cleaned_df.drop(columns=['Cabin'], errors='ignore')

# 3. Convert categorical columns to category dtype (optional, for analysis)
cleaned_df['Sex'] = cleaned_df['Sex'].astype('category')
cleaned_df['Embarked'] = cleaned_df['Embarked'].astype('category')
cleaned_df['Pclass'] = cleaned_df['Pclass'].astype('category')

print("Transformation complete. Sample data:")

print("\nMissing values after transformation:")
print(cleaned_df.isnull().sum())
```

# VISUALIZATION

```python

print("\n# VISUALIZATION 1: SURVIVAL RATE BY GENDER")
plt.figure(figsize=(6, 4))
sns.barplot(data=cleaned_df, x="Sex", y="Survived", errorbar=None)
plt.title("Survival Rate by Gender")
plt.xlabel("Gender")
plt.ylabel("Survival Rate")
plt.show()

print("\n# VISUALIZATION 2: DISTRIBUTION OF AGES")
plt.figure(figsize=(6, 4))
plt.hist(cleaned_df['Age'], bins=10)
plt.title("Distribution of Ages")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()

print("\n# VISUALIZATION 3: AGE VS FARE")
plt.figure(figsize=(6, 4))
plt.scatter(cleaned_df['Age'], cleaned_df['Fare'])
plt.title("Age vs. Fare")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()

print("\n# VISUALIZATION 4: AGE DISTRIBUTION BY CLASS")
plt.figure(figsize=(6, 4))
sns.boxplot(data=cleaned_df, x='Pclass', y='Age')
plt.title("Age Distribution by Class")
plt.xlabel("Passenger Class")
plt.ylabel("Age")
plt.show()
```
